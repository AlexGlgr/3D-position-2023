```python
import asyncio

class SetInterval:
    def __init__(self, interval, fn):
        self.interval = interval
        self.fn = fn
        self.is_running = False
        self.task = None

    async def run(self):
        self.is_running = True
        while self.is_running:
            await asyncio.sleep(self.interval)
            self.fn()

    def start(self):
        self.task = asyncio.create_task(self.run())

    def stop(self):
        self.is_running = False

# Пример использования
def helloWorld_1():
    print("Hello world!---1")
def helloWorld_2():
    print("Hello world!---2")

interval_1 = SetInterval(1,   helloWorld_1)
interval_1.start()

interval_2 = SetInterval(0.25, helloWorld_2)
interval_2.start()

# Ждем 10 секунд, затем останавливаем выполнение
await asyncio.sleep(5)
interval_1.stop()
interval_2.stop()
```

```python
import asyncio
import asyncio
import time

class SetInterval:
    def __init__(self, interval, fn):
        self.interval = interval
        self.fn = fn
        self.is_running = False
        self.task = None
        self.last_execution_time = None

    async def run(self):
        self.is_running = True
        while self.is_running:
            await asyncio.sleep(self.interval)
            self.calculate_elapsed_time()
            self.fn()

    def calculate_elapsed_time(self):
        current_time = time.time()
        if self.last_execution_time is not None:
            elapsed_time = current_time - self.last_execution_time
            print(f"Прошло времени: {elapsed_time} секунд")
        self.last_execution_time = current_time

    def start(self):
        self.task = asyncio.create_task(self.run())

    def stop(self):
        self.is_running = False

# Пример использования
def helloWorld_1():
    print("Hello world!---1")
def helloWorld_2():
    print("Hello world!---2")

#interval_1 = SetInterval(0.025,   helloWorld_1)
#interval_1.start()

interval_2 = SetInterval(0.05, helloWorld_2)
interval_2.start()

# Ждем 10 секунд, затем останавливаем выполнение
await asyncio.sleep(2)
#interval_1.stop()
interval_2.stop()

```
