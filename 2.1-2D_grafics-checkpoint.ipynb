```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as an
from datetime import datetime
import random

%matplotlib widget
```

```python
fig = plt.figure(figsize=(8,8))
ax  = fig.subplots(nrows=4, ncols=1)
```

```python
x=[]
y=[]
i=0
cur_number=0;

def UpdateGraph(frame):
    #cur_datetime=datetime.now()
    global cur_number
    cur_number=cur_number+1
    x.append(cur_number)
    y.append(random.randint(5,10))
    
    if len(x)>20 :
        x.pop(0)
        y.pop(0)

    for axi in ax:
        axi.clear()
        axi.plot(x, y)
    
fig.suptitle('ГРАФИК ПОКАЗАНИЙ АКСЕЛЕРОМЕТРА')
ax[3].set_xlabel('ВРЕМЯ')
ax[3].set_ylabel('ЗНАЧЕНИЯ АКСЕЛЕРОМЕТРА')

anim = an.FuncAnimation(fig=fig, func=UpdateGraph, frames=100, interval=500)

plt.show()
```
