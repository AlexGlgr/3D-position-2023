```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as an
from datetime import datetime
import random

%matplotlib widget
```

```python
fig = plt.figure(figsize=(8,8))
ax  = fig.subplots(nrows=4, ncols=1)
fig.suptitle('ГРАФИК ПОКАЗАНИЙ АКСЕЛЕРОМЕТРА')
```

```python
x=[]
y=[]
y2=[]
y3=[]
i=0
cur_number=0;

def UpdateGraph(frame):
    #cur_datetime=datetime.now()
    global cur_number
    cur_number=cur_number+1/8
    x.append(cur_number)
    y.append(random.randint(0,1)/10+ np.sin(cur_number+1))
    y2.append(random.randint(0,1)/10+np.sin(cur_number+2))
    y3.append(random.randint(0,1)/10+np.sin(cur_number+3))
    
    if len(x)>160 :
        x.pop(0)
        y.pop(0)

    for axi in ax:
        axi.clear()
        axi.plot(x, y, x, y2)
    
    ax[3].set_xlabel('ВРЕМЯ')
    ax[3].set_ylabel('ЗНАЧЕНИЯ ГИРОСКОПА')
    


anim = an.FuncAnimation(fig=fig, func=UpdateGraph, frames=5, interval=1000)

plt.show()
```
